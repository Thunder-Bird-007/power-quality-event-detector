```python
import numpy as np
import matplotlib.pyplot as plt

f0 = 50
Fs = 1000
T = 1.0
Vm = 1.0
```

```python
N = int(Fs * T)
t = np.arange(N) / Fs
v = Vm * np.sin(2 * np.pi * f0 * t)

show_T = 0.1
show_N = int(Fs * show_T)

plt.figure()
plt.plot(t[:show_N], v[:show_N])
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Clean 50 Hz Sine Wave (first 0.1 s)")
plt.grid(True)
plt.show()
```
